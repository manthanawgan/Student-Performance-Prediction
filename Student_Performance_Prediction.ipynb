```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('/content/student_performance_dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
df.isnull().sum()
```

```python
if 'Student_ID' in df.columns:
  df.drop('Student_ID', axis = 1, inplace = True)
  print("Dropped Student_ID column")


categorical_cols = ['Gender','Parental_Education_Level','Internet_Access_at_Home','Extracurricular_Activities','Pass_Fail']
for cols in categorical_cols:
  df[cols] = df[cols].astype('category')

print(df.dtypes)
```

```python
df.head()
```

```python
#one-hot enccoding the categorical data

df_encode = pd.get_dummies(df, drop_first=True)
print("encode features: ")
display(df.head())
display(df_encode.head())
display(df_encode.shape)

```
